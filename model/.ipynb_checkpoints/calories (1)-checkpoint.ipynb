```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dailyActivity_merged.csv")
df
```

```python
df2=df.drop('Id',axis=1)
df2
```

```python
df2['ActivityDate']=pd.to_datetime(df2['ActivityDate'])
df2['ActivityDate']
```

```python
df2.isnull().sum()
```

```python
df2=df2.set_index('ActivityDate')
df2.head()
```

```python
y=df2['Calories'].resample('D').mean()
y
```

```python
!pip install pmdarima
```

```python
import pmdarima
from pmdarima.arima import auto_arima
stepwise_model=auto_arima(y,start_p=1,start_q=1,max_p=5,max_q=5,m=12,start_P=0,seasonal=True,d=1,D=1,trace=True,error_action='ignore',suppress_warnings=True,stepwise=True)
```

```python
stepwise_model.aic()
```

```python
stepwise_model.fit(y)
```

```python
future_forecast=stepwise_model.predict(n_periods=8)
future_forecast
```

```python
import pickle
with open('calories.pkl', 'wb') as f:
    pickle.dump(stepwise_model, f)

```
